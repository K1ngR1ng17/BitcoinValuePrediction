```python
bc.head()
```

```python
bc["target"] = bc.close.shift(-1) # сдвигаем колонку close, чтобы в колонке target была цена след. дня
bc.head()
# pandas.Series == колонка|
```

```python
bc["open_close"] = bc.open - bc.close # разница между открытием и закрытием торгов
bc["high_low"] = bc.high - bc.low # разница между мин. и макс. ценой за день
```

```python
# на основе исторических данных
# хочу узнать среднюю цену открытия за последнюю неделю
# или например максимальный объем торгов за последние три дня

# rolling window = скользящее окно
bc["open_7d_mean"] = bc.open.rolling(window=7).mean()
bc["volume_3d_max"] = bc["Volume USD"].rolling(window=3).max()
bc["close_1d"] = bc.close.shift(1) # цена закрытия вчера
bc["close_2d"] = bc.close.shift(2) # цена закрытия позавчера
bc["close_3d"] = bc.close.shift(3) # цена закрытия поза-позавчера
```

```python
# Добавим колонки на основе даты
bc["weekday"] = bc.date.dt.weekday # получаем день недели из даты
bc["month"] = bc.date.dt.month
bc["year"] = bc.date.dt.year
```

```python
bc.dropna(inplace=True) # Удалим строки в которых есть пропуски данных
# Преобразовать категории
df = pd.get_dummies(bc, columns=["weekday", "month", "year"])
# Удалим лишние колонки
df.drop("symbol", axis=1, inplace=True) # удалим колонку symbol
df.drop("date", axis=1, inplace=True) # удалим колонку date
df.head()
```

```python
# Х - это то, на основе чего мы делаем предсказание
# y - это то, что мы предсказываем
X = df.drop("target", axis=1) # в Х попадет все, кроме колонки target, сама колонка при этом останется
y = df.target
```

```python
# Делим выборку на тренировочную и тестовую
# Тренировочная (обучающая / train) - выборка на которой обучается модель, "учебник" - модель учится, знает X и y
# Тестовая (проверочная / test) - выборка на которой мы проверим качество модели, "экзамен", мы просим модель зная X предсказать y (а мы сравним с правильными ответами)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# X_train, y_train - для обучения модели
# X_test - для проверки модели, вернет ли модель результаты похожие на y_test
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor() # создаем модель
# Пример плохих параметров: n_estimators=1, max_depth=2
model.fit(X_train, y_train) # обучаем
y_pred = model.predict(X_test) # просим сделать предсказание для X_test
# сравниваем y_pred и y_test
```

```python
from sklearn.metrics import mean_absolute_error, r2_score, max_error
print("MAE = ", mean_absolute_error(y_test, y_pred)) # в среднем модель ошибалась на $259 (предсказывая цену на след. день)
print("R2 = ", r2_score(y_test, y_pred)) # коэф. детерминации
print("Max = ", max_error(y_test, y_pred)) # макс. ошибка
```

```python
# Какие колонки полезны?
pd.DataFrame(data=[model.feature_importances_], columns=X.columns).T
```

```python
full_prediction = model.predict(X) # просим модель предсказать
error = y - full_prediction # покажем ошибку которую допустила модель на каждый день
plt.plot(bc.date, error, label="Absolute Error")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression() # создаем модель
model.fit(X_train, y_train) # обучаем
y_pred = model.predict(X_test) # просим сделать предсказание для X_test
print("MAE = ", mean_absolute_error(y_test, y_pred)) # в среднем модель ошибалась на $259 (предсказывая цену на след. день)
print("R2 = ", r2_score(y_test, y_pred)) # коэф. детерминации
print("Max = ", max_error(y_test, y_pred)) # макс. ошибка
full_prediction = model.predict(X) # просим модель предсказать
error = y - full_prediction # покажем ошибку которую допустила модель на каждый день
plt.plot(bc.date, error, label="Absolute Error")
plt.legend()
plt.show()
```

```python
# Резкое изменение цены
last_year = bc[bc['date'] > '2021-01-01'] # возьмем данные за 2021г.
plt.plot(last_year.date, last_year.open, label="open price") # отобразим их на графике
# найдем все примеры где цена менялась на 5000 за период 7 дней
large_change = (last_year["close"].shift(7) - last_year["close"]) > 5000 

# отобразим эти примеры в виде вертикальных красных линий
for d in list(last_year[large_change].date):
  plt.axvline(d, color='red',alpha=0.5)
plt.show()
```
